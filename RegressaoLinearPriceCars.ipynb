```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
carros = pd.read_csv('/content/C├│pia de CarPrice_Assignment.csv')
```

```python
carros.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# prompt: gerar todos os graficos de correla;áo entre as variaveis numericas com a variavel price

import matplotlib.pyplot as plt

for col in carros.select_dtypes(include=np.number):
  if col != "price":
    sns.regplot(x=col, y="price", data=carros)
    plt.show()
```

```python
le.fit(carros.CarName)
carros.CarName = le.transform(carros.CarName)

le.fit(carros.fueltype)
carros.fueltype = le.transform(carros.fueltype)

le.fit(carros.aspiration)
carros.aspiration = le.transform(carros.aspiration)

le.fit(carros.doornumber)
carros.doornumber = le.transform(carros.doornumber)

le.fit(carros.carbody)
carros.carbody = le.transform(carros.carbody)

le.fit(carros.drivewheel)
carros.drivewheel= le.transform(carros.drivewheel)

le.fit(carros.enginelocation)
carros.enginelocation= le.transform(carros.enginelocation)

le.fit(carros.fuelsystem)
carros.fuelsystem= le.transform(carros.fuelsystem)

le.fit(carros.enginetype)
carros.enginetype= le.transform(carros.enginetype)

le.fit(carros.cylindernumber)
carros.cylindernumber= le.transform(carros.cylindernumber)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

carros_norm = pd.DataFrame(scaler.fit_transform(carros), index = carros.index, columns = carros.columns)
```

```python
carros.info()
```

```python
carros_norm.head()
```

```python
carros.describe()
```

```python
import statsmodels.formula.api as smf
```

```python
function = 'price~CarName+carbody+enginelocation+carwidth+carheight+curbweight+enginesize+stroke+peakrpm'
model = smf.ols(formula = function, data=carros_norm).fit()
print(model.summary())
```

```python
x = carros_norm[["CarName","carbody", 'enginelocation', 'carwidth', 'carheight' ,'curbweight', 'enginesize', 'stroke', 'peakrpm']]
y = carros_norm[["price"]]
```

```python
carros_norm.columns
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y , test_size = 0.20, random_state = 42)

lr = LinearRegression()

lr.fit(x,y)
```

```python
r_sqr = lr.score(x,y)
r_sqr
```

```python
y_pred_treino = lr.predict(x_treino)
print("MAE", metrics.mean_absolute_error(y_treino, y_pred_treino))
print("MSE", metrics.mean_squared_error(y_treino, y_pred_treino))
print("MAE", np.sqrt(metrics.mean_squared_error(y_treino, y_pred_treino)))
```

```python
y_pred_teste = lr.predict(x_teste)
print("MAE", metrics.mean_absolute_error(y_teste, y_pred_teste))
print("MSE", metrics.mean_squared_error(y_teste, y_pred_teste))
print("MAE", np.sqrt(metrics.mean_squared_error(y_teste, y_pred_teste)))
```
